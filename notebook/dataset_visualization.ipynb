# Eating Disorder Perception - Dataset Visualization

This notebook provides comprehensive visualizations to explore the **Perception_EatingDisorder** variable across different demographic factors including gender, age range, current city, education level, employment status, and marital status.

## Objectives:
1. Understand the distribution of eating disorder perception across different demographics
2. Identify patterns and trends in the data
3. Create interactive visualizations for deeper insights
4. Perform statistical analysis of relationships between variables

---

## 1. Import Required Libraries

```python
# Import necessary libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import warnings

# Configure plotting settings
plt.style.use('default')
sns.set_palette("husl")
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("Libraries imported successfully!")
print("Available plotting libraries: matplotlib, seaborn, plotly")
print("Data manipulation: pandas, numpy")
```

## 2. Load and Explore Dataset

```python
# Load the dataset - Updated path
df = pd.read_csv('Dataset.csv')

# Display basic information about the dataset
print("=== DATASET OVERVIEW ===")
print(f"Dataset shape: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

print("\n=== COLUMN NAMES ===")
print("Columns in the dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print("\n=== TARGET VARIABLE ANALYSIS ===")
target_col = 'Perception_EatingDisorder'
print(f"Target variable: {target_col}")
print(f"Unique values: {df[target_col].unique()}")
print(f"Value counts:")
print(df[target_col].value_counts())
print(f"Percentage distribution:")
print(df[target_col].value_counts(normalize=True).round(4) * 100)
```

```python
# Explore key demographic variables
demographic_vars = ['Gender', 'Age_Range', 'current_city', 'Current_EducationLevel', 
                    'Employment_Status', 'Marital_Status']

print("=== DEMOGRAPHIC VARIABLES ANALYSIS ===")
for var in demographic_vars:
    if var in df.columns:
        print(f"\n{var.upper()}:")
        print(f"  Unique values ({df[var].nunique()}): {df[var].unique()}")
        print(f"  Value counts:")
        for val, count in df[var].value_counts().items():
            percentage = (count / len(df)) * 100
            print(f"    {val}: {count} ({percentage:.1f}%)")
    else:
        print(f"\n{var}: NOT FOUND in dataset")

# Display first few rows
print("\n=== FIRST 5 ROWS ===")
display(df.head())
```

## 3. Data Preprocessing and Cleaning

```python
# Check for missing values
print("=== MISSING VALUES ANALYSIS ===")
missing_data = df.isnull().sum()
missing_percentage = (missing_data / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing_data.index,
    'Missing_Count': missing_data.values,
    'Missing_Percentage': missing_percentage.values
}).sort_values('Missing_Count', ascending=False)

print("Columns with missing values:")
print(missing_df[missing_df['Missing_Count'] > 0])

if missing_df['Missing_Count'].sum() == 0:
    print("✓ No missing values found in the dataset!")

# Create a copy for visualization
df_viz = df.copy()

# Check the actual values in the target variable
print(f"\n=== ORIGINAL TARGET VARIABLE VALUES ===")
print(f"Unique values in Perception_EatingDisorder:")
print(df['Perception_EatingDisorder'].value_counts())

# Standardize the target variable for consistency - Complete fix
df_viz['Eating_Disorder_Perception'] = df_viz['Perception_EatingDisorder'].apply(
    lambda x: 'Yes' if 'Yes' in str(x) else 'No'
)

print(f"\n=== CLEANED TARGET VARIABLE ===")
print("New target variable: 'Eating_Disorder_Perception'")
print(df_viz['Eating_Disorder_Perception'].value_counts())

# Define key variables for analysis
target_var = 'Eating_Disorder_Perception'
demographic_vars = ['Gender', 'Age_Range', 'current_city', 'Current_EducationLevel', 
                   'Employment_Status', 'Marital_Status']

print(f"\n=== VARIABLES FOR ANALYSIS ===")
print(f"Target variable: {target_var}")
print(f"Demographic variables: {demographic_vars}")
print("Data ready for visualization!")
```

## 4. Visualize Perception by Gender

```python
# IMPROVED GENDER ANALYSIS - Focus on Percentages
print("=== EATING DISORDER PERCEPTION BY GENDER ===")

# Calculate percentages
gender_perception_pct = pd.crosstab(df_viz['Gender'], df_viz[target_var], normalize='index') * 100
gender_counts = pd.crosstab(df_viz['Gender'], df_viz[target_var])

print("\nPercentage breakdown by gender:")
for gender in gender_perception_pct.index:
    yes_pct = gender_perception_pct.loc[gender, 'Yes']
    no_pct = gender_perception_pct.loc[gender, 'No']
    total = gender_counts.loc[gender].sum()
    print(f"{gender}: Yes = {yes_pct:.1f}%, No = {no_pct:.1f}% (n={total})")

# Create focused percentage visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Eating Disorder Perception by Gender - Percentage Analysis', fontsize=16, fontweight='bold')

# 1. Percentage bar chart
gender_perception_pct.plot(kind='bar', ax=axes[0,0], 
                          color=['#FF6B6B', '#4ECDC4'], 
                          width=0.7)
axes[0,0].set_title('Percentage Distribution by Gender')
axes[0,0].set_ylabel('Percentage (%)')
axes[0,0].set_xlabel('Gender')
axes[0,0].legend(['No', 'Yes'], title='Eating Disorder Perception')
axes[0,0].tick_params(axis='x', rotation=0)
# Add percentage labels
for container in axes[0,0].containers:
    axes[0,0].bar_label(container, fmt='%.1f%%')

# 2. Side-by-side percentage comparison
x = range(len(gender_perception_pct.index))
width = 0.35
axes[1,0].bar([i - width/2 for i in x], gender_perception_pct['No'], width, 
             label='No', color='#FF6B6B', alpha=0.8)
axes[1,0].bar([i + width/2 for i in x], gender_perception_pct['Yes'], width, 
             label='Yes', color='#4ECDC4', alpha=0.8)
axes[1,0].set_xlabel('Gender')
axes[1,0].set_ylabel('Percentage (%)')
axes[1,0].set_title('Side-by-Side Percentage Comparison')
axes[1,0].set_xticks(x)
axes[1,0].set_xticklabels(gender_perception_pct.index)
axes[1,0].legend()
# Add percentage labels
for i, gender in enumerate(gender_perception_pct.index):
    axes[1,0].text(i - width/2, gender_perception_pct.loc[gender, 'No'] + 1, 
                  f"{gender_perception_pct.loc[gender, 'No']:.1f}%", 
                  ha='center', va='bottom')
    axes[1,0].text(i + width/2, gender_perception_pct.loc[gender, 'Yes'] + 1, 
                  f"{gender_perception_pct.loc[gender, 'Yes']:.1f}%", 
                  ha='center', va='bottom')

# 3. Pie charts for each gender
for idx, gender in enumerate(gender_perception_pct.index):
    row, col = (0, 1) if idx == 0 else (1, 1)
    values = [gender_perception_pct.loc[gender, 'No'], gender_perception_pct.loc[gender, 'Yes']]
    colors = ['#FF6B6B', '#4ECDC4']
    axes[row, col].pie(values, labels=['No', 'Yes'], autopct='%1.1f%%', 
                      colors=colors, startangle=90)
    axes[row, col].set_title(f'{gender} Distribution\n(n={gender_counts.loc[gender].sum()})')

plt.tight_layout()
plt.show()

# Statistical summary
print(f"\n=== DETAILED STATISTICAL SUMMARY ===")
for gender in df_viz['Gender'].unique():
    gender_data = df_viz[df_viz['Gender'] == gender]
    yes_count = sum(gender_data[target_var] == 'Yes')
    no_count = sum(gender_data[target_var] == 'No') 
    total_count = len(gender_data)
    yes_percentage = (yes_count / total_count) * 100
    no_percentage = (no_count / total_count) * 100
    print(f"\n{gender.upper()}:")
    print(f"  • Yes (believes has eating disorder): {yes_count} ({yes_percentage:.1f}%)")
    print(f"  • No (doesn't believe has eating disorder): {no_count} ({no_percentage:.1f}%)")
    print(f"  • Total participants: {total_count}")
```

## 5. Visualize Perception by Age Range

```python
# IMPROVED AGE RANGE ANALYSIS - Focus on Percentages
print("=== EATING DISORDER PERCEPTION BY AGE RANGE ===")

# Calculate percentages
age_perception_pct = pd.crosstab(df_viz['Age_Range'], df_viz[target_var], normalize='index') * 100
age_counts = pd.crosstab(df_viz['Age_Range'], df_viz[target_var])

print("\nPercentage breakdown by age range:")
for age in age_perception_pct.index:
    yes_pct = age_perception_pct.loc[age, 'Yes']
    no_pct = age_perception_pct.loc[age, 'No']
    total = age_counts.loc[age].sum()
    print(f"{age}: Yes = {yes_pct:.1f}%, No = {no_pct:.1f}% (n={total})")

# Create focused percentage visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Eating Disorder Perception by Age Range - Percentage Analysis', fontsize=16, fontweight='bold')

# 1. Percentage bar chart
age_perception_pct.plot(kind='bar', ax=axes[0,0], 
                       color=['#FF6B6B', '#4ECDC4'], 
                       width=0.7)
axes[0,0].set_title('Percentage Distribution by Age Range')
axes[0,0].set_ylabel('Percentage (%)')
axes[0,0].set_xlabel('Age Range')
axes[0,0].legend(['No', 'Yes'], title='Eating Disorder Perception')
axes[0,0].tick_params(axis='x', rotation=0)
# Add percentage labels
for container in axes[0,0].containers:
    axes[0,0].bar_label(container, fmt='%.1f%%')

# 2. Side-by-side percentage comparison
x = range(len(age_perception_pct.index))
width = 0.35
axes[0,1].bar([i - width/2 for i in x], age_perception_pct['No'], width, 
             label='No', color='#FF6B6B', alpha=0.8)
axes[0,1].bar([i + width/2 for i in x], age_perception_pct['Yes'], width, 
             label='Yes', color='#4ECDC4', alpha=0.8)
axes[0,1].set_xlabel('Age Range')
axes[0,1].set_ylabel('Percentage (%)')
axes[0,1].set_title('Side-by-Side Percentage Comparison')
axes[0,1].set_xticks(x)
axes[0,1].set_xticklabels(age_perception_pct.index)
axes[0,1].legend()
# Add percentage labels
for i, age in enumerate(age_perception_pct.index):
    axes[0,1].text(i - width/2, age_perception_pct.loc[age, 'No'] + 1, 
                  f"{age_perception_pct.loc[age, 'No']:.1f}%", 
                  ha='center', va='bottom')
    axes[0,1].text(i + width/2, age_perception_pct.loc[age, 'Yes'] + 1, 
                  f"{age_perception_pct.loc[age, 'Yes']:.1f}%", 
                  ha='center', va='bottom')

# 3. "Yes" percentage comparison
axes[1,0].bar(age_perception_pct.index, age_perception_pct['Yes'], 
             color='#4ECDC4', alpha=0.8, width=0.6)
axes[1,0].set_title('Percentage Saying "Yes" by Age Range')
axes[1,0].set_ylabel('Percentage (%)')
axes[1,0].set_xlabel('Age Range')
# Add percentage labels
for i, age in enumerate(age_perception_pct.index):
    axes[1,0].text(i, age_perception_pct.loc[age, 'Yes'] + 1, 
                  f"{age_perception_pct.loc[age, 'Yes']:.1f}%", 
                  ha='center', va='bottom')

# 4. Sample size and percentage combined
ax2 = axes[1,1].twinx()
bars1 = axes[1,1].bar(age_perception_pct.index, 
                     [age_counts.loc[age].sum() for age in age_perception_pct.index], 
                     alpha=0.6, color='lightgray', label='Sample Size')
line1 = ax2.plot(age_perception_pct.index, age_perception_pct['Yes'], 
                'ro-', linewidth=2, markersize=8, label='Yes %')
axes[1,1].set_xlabel('Age Range')
axes[1,1].set_ylabel('Sample Size', color='gray')
ax2.set_ylabel('Percentage Saying Yes (%)', color='red')
axes[1,1].set_title('Sample Size vs Percentage Analysis')
# Add labels
for i, age in enumerate(age_perception_pct.index):
    axes[1,1].text(i, age_counts.loc[age].sum() + 5, 
                  f"n={age_counts.loc[age].sum()}", 
                  ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Statistical summary
print(f"\n=== DETAILED STATISTICAL SUMMARY ===")
for age_range in df_viz['Age_Range'].unique():
    age_data = df_viz[df_viz['Age_Range'] == age_range]
    yes_count = sum(age_data[target_var] == 'Yes')
    no_count = sum(age_data[target_var] == 'No')
    total_count = len(age_data)
    yes_percentage = (yes_count / total_count) * 100
    no_percentage = (no_count / total_count) * 100
    print(f"\n{age_range.upper()}:")
    print(f"  • Yes (believes has eating disorder): {yes_count} ({yes_percentage:.1f}%)")
    print(f"  • No (doesn't believe has eating disorder): {no_count} ({no_percentage:.1f}%)")
    print(f"  • Total participants: {total_count}")
```

## 6. Visualize Perception by Current City

```python
# IMPROVED CITY ANALYSIS - Focus on Percentages
print("=== EATING DISORDER PERCEPTION BY CURRENT CITY ===")

# Calculate percentages
city_perception_pct = pd.crosstab(df_viz['current_city'], df_viz[target_var], normalize='index') * 100
city_counts = pd.crosstab(df_viz['current_city'], df_viz[target_var])

print("\nPercentage breakdown by city:")
for city in city_perception_pct.index:
    yes_pct = city_perception_pct.loc[city, 'Yes']
    no_pct = city_perception_pct.loc[city, 'No']
    total = city_counts.loc[city].sum()
    print(f"{city}: Yes = {yes_pct:.1f}%, No = {no_pct:.1f}% (n={total})")

# Sort cities by "Yes" percentage for better visualization
city_yes_sorted = city_perception_pct['Yes'].sort_values(ascending=False)
city_order = city_yes_sorted.index.tolist()

# Create focused percentage visualizations
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Eating Disorder Perception by City - Percentage Analysis', fontsize=16, fontweight='bold')

# 1. Percentage bar chart (sorted by Yes%)
city_perception_pct_sorted = city_perception_pct.loc[city_order]
city_perception_pct_sorted.plot(kind='bar', ax=axes[0,0], 
                               color=['#FF6B6B', '#4ECDC4'], 
                               width=0.8)
axes[0,0].set_title('Percentage Distribution by City (Sorted by Yes%)')
axes[0,0].set_ylabel('Percentage (%)')
axes[0,0].set_xlabel('City')
axes[0,0].legend(['No', 'Yes'], title='Eating Disorder Perception')
axes[0,0].tick_params(axis='x', rotation=45)
# Add percentage labels
for container in axes[0,0].containers:
    axes[0,0].bar_label(container, fmt='%.1f%%')

# 2. Horizontal bar chart showing Yes percentages ranked
axes[0,1].barh(range(len(city_yes_sorted)), city_yes_sorted.values, 
              color='#4ECDC4', alpha=0.8)
axes[0,1].set_yticks(range(len(city_yes_sorted)))
axes[0,1].set_yticklabels(city_yes_sorted.index)
axes[0,1].set_title('Cities Ranked by "Yes" Percentage')
axes[0,1].set_xlabel('Percentage Saying Yes (%)')
# Add percentage labels
for i, (city, pct) in enumerate(city_yes_sorted.items()):
    sample_size = city_counts.loc[city].sum()
    axes[0,1].text(pct + 0.5, i, f"{pct:.1f}% (n={sample_size})", 
                  va='center', ha='left')

# 3. Side-by-side comparison for top cities
top_cities = city_order[:5]  # Top 5 cities by Yes%
x = range(len(top_cities))
width = 0.35
no_pcts = [city_perception_pct.loc[city, 'No'] for city in top_cities]
yes_pcts = [city_perception_pct.loc[city, 'Yes'] for city in top_cities]

axes[1,0].bar([i - width/2 for i in x], no_pcts, width, 
             label='No', color='#FF6B6B', alpha=0.8)
axes[1,0].bar([i + width/2 for i in x], yes_pcts, width, 
             label='Yes', color='#4ECDC4', alpha=0.8)
axes[1,0].set_xlabel('City')
axes[1,0].set_ylabel('Percentage (%)')
axes[1,0].set_title('Top 5 Cities - Side-by-Side Percentage Comparison')
axes[1,0].set_xticks(x)
axes[1,0].set_xticklabels([city[:8] + '...' if len(city) > 8 else city for city in top_cities])
axes[1,0].legend()
# Add percentage labels
for i, city in enumerate(top_cities):
    axes[1,0].text(i - width/2, no_pcts[i] + 1, f"{no_pcts[i]:.1f}%", 
                  ha='center', va='bottom')
    axes[1,0].text(i + width/2, yes_pcts[i] + 1, f"{yes_pcts[i]:.1f}%", 
                  ha='center', va='bottom')

# 4. Sample sizes with percentage overlay
sample_sizes = [city_counts.loc[city].sum() for city in city_order]
ax2 = axes[1,1].twinx()
bars = axes[1,1].bar(range(len(city_order)), sample_sizes, 
                    alpha=0.6, color='lightgray')
line = ax2.plot(range(len(city_order)), 
               [city_perception_pct.loc[city, 'Yes'] for city in city_order], 
               'ro-', linewidth=2, markersize=6)
axes[1,1].set_xlabel('City (Ranked by Yes%)')
axes[1,1].set_ylabel('Sample Size', color='gray')
ax2.set_ylabel('Percentage Saying Yes (%)', color='red')
axes[1,1].set_title('Sample Size vs Yes Percentage by City')
axes[1,1].set_xticks(range(len(city_order)))
axes[1,1].set_xticklabels([city[:6] + '...' if len(city) > 6 else city for city in city_order], 
                         rotation=45)

plt.tight_layout()
plt.show()

# Statistical summary
print(f"\n=== DETAILED STATISTICAL SUMMARY ===")
print("Cities ranked by eating disorder perception rate:")
for i, (city, percentage) in enumerate(city_yes_sorted.items(), 1):
    city_data = df_viz[df_viz['current_city'] == city]
    yes_count = sum(city_data[target_var] == 'Yes')
    no_count = sum(city_data[target_var] == 'No')
    total_count = len(city_data)
    no_percentage = (no_count / total_count) * 100
    print(f"\n{i}. {city.upper()}:")
    print(f"   • Yes (believes has eating disorder): {yes_count} ({percentage:.1f}%)")
    print(f"   • No (doesn't believe has eating disorder): {no_count} ({no_percentage:.1f}%)")
    print(f"   • Total participants: {total_count}")
```

## 7. Create Cross-tabulation Analysis

```python
# IMPROVED MULTI-DIMENSIONAL ANALYSIS - Focus on Percentages
print("=== COMPREHENSIVE DEMOGRAPHIC PERCENTAGE ANALYSIS ===")

# Calculate percentages for all demographic variables
demographic_percentages = {}
for var in demographic_vars:
    if var in df_viz.columns:
        var_perception_pct = pd.crosstab(df_viz[var], df_viz[target_var], normalize='index') * 100
        demographic_percentages[var] = var_perception_pct

# Create comprehensive visualization
fig, axes = plt.subplots(3, 2, figsize=(20, 18))
fig.suptitle('Eating Disorder Perception Across All Demographics - Percentage Analysis', 
             fontsize=18, fontweight='bold')

# 1. Education Level
if 'Current_EducationLevel' in demographic_percentages:
    edu_pct = demographic_percentages['Current_EducationLevel']
    edu_pct.plot(kind='bar', ax=axes[0,0], color=['#FF6B6B', '#4ECDC4'], width=0.8)
    axes[0,0].set_title('Education Level')
    axes[0,0].set_ylabel('Percentage (%)')
    axes[0,0].legend(['No', 'Yes'], title='Eating Disorder Perception')
    axes[0,0].tick_params(axis='x', rotation=45)
    # Add percentage labels for Yes only
    for i, (idx, row) in enumerate(edu_pct.iterrows()):
        axes[0,0].text(i, row['Yes'] + row['No'] + 2, f"{row['Yes']:.1f}%", 
                      ha='center', va='bottom', fontweight='bold')

# 2. Employment Status
if 'Employment_Status' in demographic_percentages:
    emp_pct = demographic_percentages['Employment_Status']
    emp_pct.plot(kind='bar', ax=axes[0,1], color=['#FF6B6B', '#4ECDC4'], width=0.8)
    axes[0,1].set_title('Employment Status')
    axes[0,1].set_ylabel('Percentage (%)')
    axes[0,1].legend(['No', 'Yes'], title='Eating Disorder Perception')
    axes[0,1].tick_params(axis='x', rotation=45)
    # Add percentage labels for Yes only
    for i, (idx, row) in enumerate(emp_pct.iterrows()):
        axes[0,1].text(i, row['Yes'] + row['No'] + 2, f"{row['Yes']:.1f}%", 
                      ha='center', va='bottom', fontweight='bold')

# 3. Marital Status
if 'Marital_Status' in demographic_percentages:
    mar_pct = demographic_percentages['Marital_Status']
    mar_pct.plot(kind='bar', ax=axes[1,0], color=['#FF6B6B', '#4ECDC4'], width=0.8)
    axes[1,0].set_title('Marital Status')
    axes[1,0].set_ylabel('Percentage (%)')
    axes[1,0].legend(['No', 'Yes'], title='Eating Disorder Perception')
    axes[1,0].tick_params(axis='x', rotation=45)
    # Add percentage labels for Yes only
    for i, (idx, row) in enumerate(mar_pct.iterrows()):
        axes[1,0].text(i, row['Yes'] + row['No'] + 2, f"{row['Yes']:.1f}%", 
                      ha='center', va='bottom', fontweight='bold')

# 4. Gender vs Age Range Combined
multi_crosstab_pct = pd.crosstab([df_viz['Gender'], df_viz['Age_Range']], 
                                df_viz[target_var], normalize='index') * 100
multi_crosstab_pct.plot(kind='bar', ax=axes[1,1], color=['#FF6B6B', '#4ECDC4'], width=0.8)
axes[1,1].set_title('Gender × Age Range Combined')
axes[1,1].set_ylabel('Percentage (%)')
axes[1,1].legend(['No', 'Yes'], title='Eating Disorder Perception')
axes[1,1].tick_params(axis='x', rotation=45)
# Add percentage labels for Yes only
for i, (idx, row) in enumerate(multi_crosstab_pct.iterrows()):
    axes[1,1].text(i, row['Yes'] + row['No'] + 2, f"{row['Yes']:.1f}%", 
                  ha='center', va='bottom', fontweight='bold')

# 5. Summary of "Yes" percentages across all demographics
yes_percentages_summary = []
categories_summary = []
sample_sizes_summary = []

for var in demographic_vars:
    if var in df_viz.columns:
        for category in df_viz[var].unique():
            subset = df_viz[df_viz[var] == category]
            yes_pct = (subset[target_var] == 'Yes').mean() * 100
            yes_percentages_summary.append(yes_pct)
            categories_summary.append(f"{var}:\n{category}"[:20] + "..." if len(f"{var}: {category}") > 20 else f"{var}:\n{category}")
            sample_sizes_summary.append(len(subset))

# Sort by percentage
sorted_indices = sorted(range(len(yes_percentages_summary)), 
                       key=lambda i: yes_percentages_summary[i], reverse=True)

top_10_indices = sorted_indices[:10]
top_10_pcts = [yes_percentages_summary[i] for i in top_10_indices]
top_10_cats = [categories_summary[i] for i in top_10_indices]
top_10_sizes = [sample_sizes_summary[i] for i in top_10_indices]

axes[2,0].barh(range(len(top_10_pcts)), top_10_pcts, color='#4ECDC4', alpha=0.8)
axes[2,0].set_yticks(range(len(top_10_pcts)))
axes[2,0].set_yticklabels(top_10_cats, fontsize=8)
axes[2,0].set_title('Top 10: Highest "Yes" Percentages')
axes[2,0].set_xlabel('Percentage Saying Yes (%)')
# Add labels
for i, (pct, size) in enumerate(zip(top_10_pcts, top_10_sizes)):
    axes[2,0].text(pct + 0.5, i, f"{pct:.1f}% (n={size})", 
                  va='center', ha='left', fontsize=8)

# 6. Bottom 10 percentages
bottom_10_indices = sorted_indices[-10:]
bottom_10_pcts = [yes_percentages_summary[i] for i in bottom_10_indices]
bottom_10_cats = [categories_summary[i] for i in bottom_10_indices]
bottom_10_sizes = [sample_sizes_summary[i] for i in bottom_10_indices]

axes[2,1].barh(range(len(bottom_10_pcts)), bottom_10_pcts, color='#FF6B6B', alpha=0.8)
axes[2,1].set_yticks(range(len(bottom_10_pcts)))
axes[2,1].set_yticklabels(bottom_10_cats, fontsize=8)
axes[2,1].set_title('Bottom 10: Lowest "Yes" Percentages')
axes[2,1].set_xlabel('Percentage Saying Yes (%)')
# Add labels
for i, (pct, size) in enumerate(zip(bottom_10_pcts, bottom_10_sizes)):
    axes[2,1].text(pct + 0.5, i, f"{pct:.1f}% (n={size})", 
                  va='center', ha='left', fontsize=8)

plt.tight_layout()
plt.show()

# Detailed statistical summary
print("\n=== COMPREHENSIVE DEMOGRAPHIC PERCENTAGE BREAKDOWN ===")
for var in demographic_vars:
    if var in df_viz.columns:
        print(f"\n{var.upper()}:")
        var_crosstab = pd.crosstab(df_viz[var], df_viz[target_var])
        var_pct = pd.crosstab(df_viz[var], df_viz[target_var], normalize='index') * 100
        
        for category in var_pct.index:
            yes_pct = var_pct.loc[category, 'Yes']
            no_pct = var_pct.loc[category, 'No']
            total = var_crosstab.loc[category].sum()
            yes_count = var_crosstab.loc[category, 'Yes']
            no_count = var_crosstab.loc[category, 'No']
            print(f"  • {category}:")
            print(f"    - Yes: {yes_count} ({yes_pct:.1f}%)")
            print(f"    - No: {no_count} ({no_pct:.1f}%)")
            print(f"    - Total: {total}")
```

```python
# Create heatmaps for categorical relationships
print("=== HEATMAP ANALYSIS ===")

fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('Heatmap Analysis: Eating Disorder Perception Across Demographics', 
             fontsize=18, fontweight='bold')

# 1. Gender vs City heatmap
gender_city_crosstab = pd.crosstab(df_viz['Gender'], df_viz['current_city'])
sns.heatmap(gender_city_crosstab, annot=True, fmt='d', cmap='Blues', ax=axes[0,0])
axes[0,0].set_title('Gender vs Current City Distribution')
axes[0,0].set_ylabel('Gender')

# 2. Age Range vs Education Level
age_edu_crosstab = pd.crosstab(df_viz['Age_Range'], df_viz['Current_EducationLevel'])
sns.heatmap(age_edu_crosstab, annot=True, fmt='d', cmap='Greens', ax=axes[0,1])
axes[0,1].set_title('Age Range vs Education Level Distribution')
axes[0,1].set_ylabel('Age Range')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Eating Disorder Perception Rate by Demographics
perception_matrix = pd.DataFrame()
for var in demographic_vars:
    if var in df_viz.columns:
        var_perception = pd.crosstab(df_viz[var], df_viz[target_var], normalize='index') * 100
        perception_matrix[var] = var_perception['Yes']

# Transpose and fill NaN values
perception_matrix_T = perception_matrix.T.fillna(0)
sns.heatmap(perception_matrix_T, annot=True, fmt='.1f', cmap='Reds', 
           cbar_kws={'label': 'Percentage saying "Yes" (%)'}, ax=axes[1,0])
axes[1,0].set_title('Eating Disorder Perception Rate by Category')
axes[1,0].set_ylabel('Demographic Variable')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. City vs Age Range for Eating Disorder Perception
city_age_perception = df_viz.groupby(['current_city', 'Age_Range'])[target_var].apply(
    lambda x: (x == 'Yes').mean() * 100
).unstack(fill_value=0)

sns.heatmap(city_age_perception, annot=True, fmt='.1f', cmap='YlOrRd', 
           cbar_kws={'label': 'Percentage saying "Yes" (%)'}, ax=axes[1,1])
axes[1,1].set_title('Eating Disorder Perception: City vs Age Range')
axes[1,1].set_ylabel('Current City')
axes[1,1].set_xlabel('Age Range')

plt.tight_layout()
plt.show()

# Create a summary correlation-like matrix for all categorical variables
print("\n=== ASSOCIATION ANALYSIS ===")
from scipy.stats import chi2_contingency

association_results = {}
categorical_vars = ['Gender', 'Age_Range', 'current_city', 'Current_EducationLevel', 
                   'Employment_Status', 'Marital_Status']

print("Chi-square test results (association with eating disorder perception):")
print("Variable" + " " * 20 + "Chi2 Statistic" + " " * 5 + "p-value" + " " * 10 + "Association")
print("-" * 70)

for var in categorical_vars:
    if var in df_viz.columns:
        contingency = pd.crosstab(df_viz[var], df_viz[target_var])
        chi2, p_value, dof, expected = chi2_contingency(contingency)
        
        # Determine association strength
        if p_value < 0.001:
            association = "Very Strong"
        elif p_value < 0.01:
            association = "Strong"
        elif p_value < 0.05:
            association = "Moderate"
        else:
            association = "Weak/None"
            
        association_results[var] = {'chi2': chi2, 'p_value': p_value, 'association': association}
        print(f"{var:<25} {chi2:>12.3f} {p_value:>15.4f} {association:>15}")

print(f"\nNote: p-value < 0.05 indicates significant association")
```
